```python
import pandas as pd
pd.options.display.max_rows = 60 #None #60
pd.options.display.max_columns = 60 # None #70
import numpy as np
import matplotlib.pyplot as plt
```

# Participante MD

## MD CONTROL

```python
# read csv
df_datos1 = pd.read_csv('MD_control.csv')
df_datos1
```

```python
#GRAFICA DE PULSO
fig,A = plt.subplots(1,1,figsize=(15,5))
A.plot(df_datos1['B: BVP'])
A.set_ylabel('BVP',fontsize=20)
A.set_xlabel('time',fontsize=15)
A.set_xlim(50,500000)                               ## DATO TUNEADO
A.set_ylim(30,45)                                ## DATO TUNEADO
A.set_title('MD control',fontsize=20)

# GRAFO CONDUCTANCIA
fig,B = plt.subplots(1,1,figsize=(15,5))
B.plot(df_datos1['E: Skin Cond'])
B.set_ylabel('Skin Conductance',fontsize=20)
B.set_xlabel('time',fontsize=15)
B.set_title('MD control',fontsize=20)

#GRAFO RESPIRACION
fig,C = plt.subplots(1,1,figsize=(15,5))
C.plot(df_datos1['G: Abd Resp'])
C.set_ylabel('Breathing',fontsize=20)
C.set_xlabel('time',fontsize=15)
C.set_title('MD control',fontsize=20)

plt.show()
```
